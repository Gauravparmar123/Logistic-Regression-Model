```python
# import the required libraries 

import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.scatterplot(x="Studied",y="Slept",data=data,hue="Passed")
```

```python
# Now taking Independent and Dependant variable's
x=data[["Studied","Slept"]]
y=data[["Passed"]]
```

```python
# Training  & Testing of Data ,import Tarin test & split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
# Apply Logistic Regression 
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
# Now fit our training data into our model
model.fit(x_train,y_train)
```

```python
# Check Score of Prediction
model.score(x_test,y_test)
```

```python
# Prediction on future data or new data
import numpy as np
newdata=(4.87,3.3)
newdata_array=np.array(newdata)
```

```python
# Now converting new data to 2d array
in_newdata=newdata_array.reshape(1,-1)
```

```python
in_newdata
```

```python
# Predict result
predict1=model.predict(in_newdata)
```
